# Feature Engineering

### normalization

```python
#scaling 
#as the features are not normally distributed we are using normalization 
x.describe()
```

```python
x.columns
```

```python
norm_scaler=MinMaxScaler()
norm_scaler.fit(x)  # x_min and x_max are calculated 
norm_data=norm_scaler.transform(x)
norm_data
df_norm=pd.DataFrame(norm_data,columns=x.columns)
df_norm
```

## Model Training

#### linear_regression model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
x_train.shape,y_train.shape,y_test.shape,x_test.shape
```

```python
lin_model=LinearRegression()
```

```python
lin_model.fit(x_train,y_train)
```

```python
lin_model.coef_
```

```python
lin_model.intercept_
```

```python
y_pred=lin_model.predict(x_train)
```

```python
df.head(1)
```

```python
mean_absolute_error(y_train,y_pred)
```

```python
#model evaluation function
def model_eval(model,x,y):
    y_pred=model.predict(x)
    mae=mean_absolute_error(y,y_pred)
    mse=mean_squared_error(y,y_pred)
    R2=r2_score(y,y_pred)
    print(f"mean squared error={mse}")
    print()
    print(f"mean absolute error={mae}")
    print()
    print(f"r 2 score={R2}")
    
    return "evaluation done"
```

```python
#model evaluation on training data 
print("#"*50)
print()
print("evaluation of training data")
model_eval(lin_model,x_train,y_train)
print("#"*50)
print()
print("evaluation of testing data")
model_eval(lin_model,x_test,y_test)
```

#### KNN REGRESSOR MODEL

```python
##KNN REGRESSOR MODEL 
knn_model=KNeighborsRegressor(n_neighbors=5,p=2)
knn_model.fit(x,y_train)  ##fitting training data
```

```python
model_eval(knn_model,x_train,y_train)
```

```python

print("#"*50)
print()
print("evaluation of training data")
model_eval(knn_model,x_train,y_train)
print("#"*50)
print()
print("evaluation of testing data")
model_eval(knn_model,x_test,y_test)
```

#### KNN model with norm data

```python
#knn_model training with normalised  data
x_train,x_test,y_train,y_test=train_test_split(df_norm,y,random_state=5,test_size=0.3)
```

```python
knn_model.fit(x_train,y_train)
```

```python
#model evaluation
print("#"*50)
print()
print("evaluation of training data")
model_eval(knn_model,x_train,y_train)
print("#"*50)
print()
print("evaluation of testing data")
model_eval(knn_model,x_test,y_test)
```
